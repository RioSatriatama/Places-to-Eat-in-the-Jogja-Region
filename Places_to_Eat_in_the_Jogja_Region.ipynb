```python
# Import Library yang akan di pakai
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Membaca file CSV
df = pd.read_csv('/content/places_to_eat_in_the_jogja_region.csv')

# Menampilkan informasi dasar tentang dataset
print(df.info())
print("\
Lima baris pertama dari dataset:")
print(df.head())
```

Di dalam Dataset tersebut berisi tentang informasi 291 restoran di wilayah Jogja dengan berbagai detail seperti: nama, jenis makanan, lokasi, harga, rating, dan lainnya.

```python
# Visualisasi distribusi harga rata-rata makanan
plt.figure(figsize=(10, 6))
sns.histplot(df['Harga Rata-Rata Makanan di Toko (Rp)'], bins=20, kde=True)
plt.title('Distribusi Harga Rata-Rata Makanan di Toko')
plt.xlabel('Harga Rata-Rata (Rp)')
plt.ylabel('Frekuensi')
plt.grid(True)
plt.show()

# Visualisasi distribusi rating toko
plt.figure(figsize=(10, 6))
sns.histplot(df['Rating Toko'], bins=10, kde=True)
plt.title('Distribusi Rating Toko')
plt.xlabel('Rating')
plt.ylabel('Frekuensi')
plt.grid(True)
plt.show()

# Visualisasi jenis preferensi makanan
plt.figure(figsize=(12, 6))
sns.countplot(y='Preferensi Makanan', data=df, order=df['Preferensi Makanan'].value_counts().index)
plt.title('Distribusi Preferensi Makanan')
plt.xlabel('Jumlah')
plt.ylabel('Preferensi Makanan')
plt.grid(True)
plt.show()
```

Penjelasan singkat:
- Grafik pertama menunjukkan distribusi harga rata-rata makanan di toko. Kita dapat melihat bagaimana harga tersebar dan apakah ada kecenderungan harga tertentu.
- Grafik kedua menunjukkan distribusi rating toko, yang membantu kita memahami bagaimana penilaian pelanggan terhadap restoran.
- Grafik ketiga menunjukkan distribusi preferensi makanan, yang memberikan gambaran tentang jenis makanan yang paling umum di restoran-restoran tersebut.

# Rangkuman dari dataset tersebut:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Membaca file CSV
df = pd.read_csv('/content/places_to_eat_in_the_jogja_region.csv')

# Ringkasan statistik
print("Ringkasan Statistik:")
print(df.describe())

# Informasi kolom
print("\
Informasi Kolom:")
print(df.info())

# Top 10 restoran dengan rating tertinggi
print("\
Top 10 Restoran dengan Rating Tertinggi:")
print(df.nlargest(10, 'Rating Toko')[['Nama Restoran', 'Rating Toko', 'Preferensi Makanan']])

# Rata-rata harga berdasarkan preferensi makanan
avg_price_by_preference = df.groupby('Preferensi Makanan')['Harga Rata-Rata Makanan di Toko (Rp)'].mean().sort_values(ascending=False)
print("\
Rata-rata Harga Berdasarkan Preferensi Makanan:")
print(avg_price_by_preference)

# Visualisasi rata-rata harga berdasarkan preferensi makanan
plt.figure(figsize=(12, 6))
avg_price_by_preference.plot(kind='bar')
plt.title('Rata-rata Harga Berdasarkan Preferensi Makanan')
plt.xlabel('Preferensi Makanan')
plt.ylabel('Rata-rata Harga (Rp)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Distribusi jenis suasana
plt.figure(figsize=(10, 6))
df['Jenis Suasana'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Distribusi Jenis Suasana')
plt.ylabel('')
plt.show()

# Korelasi antara harga dan rating
correlation = df['Harga Rata-Rata Makanan di Toko (Rp)'].corr(df['Rating Toko'])
print(f"\
Korelasi antara Harga dan Rating: {correlation:.2f}")

# Visualisasi korelasi antara harga dan rating
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Harga Rata-Rata Makanan di Toko (Rp)', y='Rating Toko', data=df)
plt.title('Korelasi antara Harga dan Rating')
plt.xlabel('Harga Rata-Rata (Rp)')
plt.ylabel('Rating Toko')
plt.show()
```
